#Распределение Рэлея
wiki: https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5_%D0%A0%D1%8D%D0%BB%D0%B5%D1%8F

$\sigma$ - параметр распределения, в этом примере везде $\sigma =1$.

Мат. ожидание распределения Релея: $\sqrt{\pi/2}*\sigma \approx 1.25...$

Дисперсия распределения Релея: $(2-\pi/2)*\sigma^2 \approx 0.43...$

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import scipy.stats as sts
```

#Построим гистограмму и график плотности вероятности для распределения Релея:

```
sigma = 1                   #зададим параметр распределния
E = np.sqrt(np.pi/2)*sigma  #вычислим мат. ожидание распределения Релея
D = (2-np.pi/2)*sigma**2    #вычислим дисперсию распределения Релея
```

```
rayleigh = sts.rayleigh() #задаем распределение Релея с параметром sigma = 1 (по умолчанию)
sample = rayleigh.rvs(1000)
x = np.linspace(0, 5)
pdf = rayleigh.pdf(x)

plt.hist(sample, density=True, label='Частоты')
plt.plot(x, pdf, label='Плотность вероятности')
plt.ylabel('Плотность вероятности / частоты')
plt.xlabel('x')
plt.legend()
```

#Для выбоки в n = 5 элементов:

```
n = 5

sample_5 = []
for i in range(1000):
  sample_mean = sum(rayleigh.rvs(n))/n #вычисляем выборочное среднее
  sample_5.append(sample_mean)
plt.hist(sample_5, density=True, label='Частоты')

norm_5 = sts.norm(E, np.sqrt(D/n)) #задаем нормальное распределение согласно центральной предельной теореме (мат. ожидание равно мат. ожиданию распределения Релея, дисперсию делим на число элементов в выборке)
pdf_5 = norm_5.pdf(x)
plt.plot(x, pdf_5, label='Апроксимация норм. распред.')
plt.ylabel('Плотность вероятности / частоты')
plt.xlabel('x')
plt.legend()
```

#Для выбоки в n = 10 элементов:

```
n = 10

sample_10 = []
for i in range(1000):
  sample_mean = sum(rayleigh.rvs(n))/n #вычисляем выборочное среднее
  sample_10.append(sample_mean)
plt.hist(sample_10, density=True, label='Частоты')

norm_10 = sts.norm(E, np.sqrt(D/n)) #задаем нормальное распределение согласно центральной предельной теореме (мат. ожидание равно мат. ожиданию распределения Релея, дисперсию делим на число элементов в выборке)
pdf_10 = norm_10.pdf(x)
plt.plot(x, pdf_10, label='Апроксимация норм. распред.')
plt.ylabel('Плотность вероятности / частоты')
plt.xlabel('x')
plt.legend()
```

#Вывод:
Нормальное распределение $\mathcal{N}(\mathbb{E}X, \frac{\mathbb{D}X}{n})$ хорошо апроксимирует выборочные средние распределения Релея. При увеличении n распределение становится "уже", но его среднее всегда равно мат. ожиданию изначального распределения
